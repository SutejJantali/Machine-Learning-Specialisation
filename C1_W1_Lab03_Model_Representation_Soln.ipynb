```python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('./deeplearning.mplstyle')
```

```python
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x train shape : {x_train}")
m = x_train.shape[0]
print(f"number of training examples is {m}")
```

```python
m = len(x_train)
print(f"Number of training examples is: {m}")
```

```python
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
plt.scatter(x_train, y_train, marker = 'x', c = 'r')
plt.title("House rental prices")
plt.ylabel("Prices in 1000's of dollars")
plt.xlabel("Size in 1000 sqft")
plt.show()
```

```python
w = 200
b = 100
print(f"w = {w}")
print(f"b = {b}")
```

```python
def compute_model_op(x_train, w, b):
    n = x_train.shape[0] #used to find length of plot size list
    f_wb = np.zeros(n)
    for i in range(n):
        f_wb[i] = w * x_train[i] + b        
    
    return f_wb
```

```python
temp_f_wb = compute_model_op(x_train, w, b,)
plt.plot(x_train, temp_f_wb, c = 'b', label = 'Our Prediction')
plt.scatter(x_train, y_train, c = 'r', marker = 'x', label = "Actual values")
plt.title("House rental prices")
plt.ylabel("Prices in 1000's of dollars")
plt.xlabel("Size in 1000 sqft")
plt.legend()
plt.show()
```
